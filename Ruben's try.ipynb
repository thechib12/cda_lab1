```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
tips = sns.load_dataset("tips")

data = pd.read_csv('data.csv')
print(data.shape)
data = data[data.simple_journal != 'Refused']
data.shape
```

```python
def num_missing(x):
    return sum(x.isnull())

print("Missing values per column:")
print(data.apply(num_missing, axis=0))

data['issuercountrycode'] = pd.Categorical(data['issuercountrycode']).codes
data['txvariantcode'] = pd.Categorical(data['txvariantcode']).codes
data['currencycode'] = pd.Categorical(data['currencycode']).codes
data['shoppercountrycode'] = pd.Categorical(data['shoppercountrycode']).codes
data['accountcode'] = pd.Categorical(data['accountcode']).codes
data['mail_id'] = pd.Categorical(data['mail_id']).codes
data['ip_id'] = pd.Categorical(data['ip_id']).codes
data['card_id'] = pd.Categorical(data['card_id']).codes
data['shopperinteraction'] = pd.Categorical(data['shopperinteraction']).codes
data['cardverificationcodesupplied'] = pd.Categorical(data['cardverificationcodesupplied']).codes
print(data.dtypes)
print(data.describe())
data.head(10)
```

```python
data.simple_journal
g1 = data[data.simple_journal == 'Settled'].sample(2000)
g2 = data[data.simple_journal == 'Chargeback']
```

```python
plt.figure()

f, axarr = plt.subplots(2, sharex=True, sharey=True)
for row in g1.itertuples(index=True):
    x, y = row.amount, row.bin
    axarr[0].scatter(x, y, alpha=0.8, c='blue', edgecolors='none', s=30)

for row in g2.itertuples(index=True):
    x, y = row.amount, row.bin
    axarr[1].scatter(x, y, alpha=0.8, c='red', edgecolors='none', s=30)
 
plt.title('Matplot scatter plot')
plt.legend(loc=2)
plt.show()
```
